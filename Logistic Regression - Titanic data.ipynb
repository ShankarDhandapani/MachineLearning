```python
titanic_dataset.drop("Cabin",axis=1, inplace=True)
```

```python
titanic_dataset.head(3)
```

```python
sex = pd.get_dummies(titanic_dataset['Sex'], drop_first=True)
sex.head(5)
```

```python
embarked = pd.get_dummies(titanic_dataset['Embarked'], drop_first=True)
embarked.head()
```

```python
pcl = pd.get_dummies(titanic_dataset['Pclass'], drop_first=True)
pcl.head()
```

```python
titanic_data = pd.concat([titanic_dataset, sex, embarked, pcl], axis=1)
titanic_data.head()
```

```python
titanic_data.drop(["Sex", "Embarked", "PassengerId", "Name", "Ticket", "Pclass"],axis=1, inplace=True)
```

```python
titanic_data.head()
```

## Test and Train Data

```python
x = titanic_data.drop("Survived", axis=1)
y = titanic_data["Survived"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel = LogisticRegression()
```

```python
print(logmodel.fit(X_train, y_train))
```

```python
predictions = logmodel.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, predictions))
```

## Accuracy Check

```python
from sklearn.metrics import confusion_matrix
```

```python
print(confusion_matrix(y_test, predictions))
```

```python
from sklearn.metrics import accuracy_score
```

```python
print("Accuracy Score : ", accuracy_score(y_test, predictions) * 100)
```
