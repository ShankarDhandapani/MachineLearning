In Logistic regression, Every values should be in number but in the below data there are some in strings 

```python
titanic_dataset.head(3)
```

So, we can generate dummies 

In the below code According to **Sex** Column there are two categories 
    1. Male
    2. Female
    
So, if the male is **1** then it represents **Male** if it is **0** then it Represents **Female**

```python
sex = pd.get_dummies(titanic_dataset['Sex'], drop_first=True)
sex.head(5)
```

In **Embarked** column there are three categories of data. They are:
    1. C - Cherbourg
    2. Q - Queenstown
    3. S - Southampton
    
Though I have removed the First column because the data can be determined by using the two data columns

```python
embarked = pd.get_dummies(titanic_dataset['Embarked'], drop_first=True)
embarked.head()
```

```python
pcl = pd.get_dummies(titanic_dataset['Pclass'], drop_first=True)
pcl.head()
```

The dummies are concatenated to the original dataset and assigned to **titanic_data** object. It is displayed below

```python
titanic_data = pd.concat([titanic_dataset, sex, embarked, pcl], axis=1)
titanic_data.head()
```

As the dummies are generated for the string data and concatenated with the original data set. The redundant data can be dropped from the dataset. The below code shows that

```python
titanic_data.drop(["Sex", "Embarked", "PassengerId", "Name", "Ticket", "Pclass"],axis=1, inplace=True)
```

```python
titanic_data.head()
```

## Test and Train Data

```python
x = titanic_data.drop("Survived", axis=1)
y = titanic_data["Survived"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel = LogisticRegression()
```

```python
print(logmodel.fit(X_train, y_train))
```

```python
predictions = logmodel.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, predictions))
```

## Accuracy Check

```python
from sklearn.metrics import confusion_matrix
```

```python
print(confusion_matrix(y_test, predictions))
```

```python
from sklearn.metrics import accuracy_score
```

```python
print("Accuracy Score : ", accuracy_score(y_test, predictions) * 100)
```
